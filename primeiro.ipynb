```python
import pandas as pd
```

```python
dados ='pedidos_narashop.csv'

df = pd.read_csv(dados, encoding= 'latin1', sep= ';')
df
```

```python
df_status_pedido_entregue= df.groupby('entregue')['status_pedido'].sum().reset_index()
df_status_pedido_entregue = df_status_pedido_entregue.sort_values(by= 'status_pedido', ascending=False)
df_status_pedido_entregue




df_furto_celular_cisp = df.groupby('cisp')['furto_celular'].sum().reset_index()
df_furto_celular_cisp = df_furto_celular_cisp.sort_values(by= 'furto_celular', ascending=False)
df_furto_celular_cisp

```
